```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Let's create a DataFrame with random data representing the "Fear Index"
date_range = pd.date_range(start='01-01-2019', end='12-31-2021', freq='B')
fear_index = np.random.randint(low=0, high=100, size=len(date_range))

df = pd.DataFrame({'Fear Index': fear_index}, index=date_range)

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Fear Index'], label='Fear Index', color='red')
plt.fill_between(df.index, df['Fear Index'], color='red', alpha=0.1)
plt.title('Stock Market Fear Index Over Time')
plt.xlabel('Date')
plt.ylabel('Fear Index')
plt.legend()
plt.grid(True)
plt.show()
```
